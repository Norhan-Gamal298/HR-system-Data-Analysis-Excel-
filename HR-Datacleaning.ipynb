```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df,x='Distance From Home',hue='Attrition',bins=30,multiple="layer",alpha=0.6)
plt.title('Distance From Home Distribution by Attrition Status')
plt.xlabel('Distance From Home ')
plt.ylabel('Count')
plt.legend(title='Attrition',labels=['No','Yes'])
plt.show()
```

## mostly all of the employees are living nearby their working place 
### so It is best to provide on-site accommodation for employees.

```python
numerical_columns=[]
for column in df.columns:
    if df[column].dtypes!='object':
        numerical_columns.append(column)
print(numerical_columns)
```

```python
correlation_matrix=df[numerical_columns].corr()
plt.figure(figsize=(15,15))
sns.heatmap(correlation_matrix,annot=True,fmt='.2f',cmap='YlGnBu',vmax=1,vmin=-1)
plt.title("correlation heatmapof HR Data")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```
